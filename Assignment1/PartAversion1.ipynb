```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
data=pd.read_csv('PartAdata.csv')
```

```python
data
```

```python
data.isnull()
```

```python
#Replace NA with the average value of the column
data.GS.fillna(data.GS.mean(),inplace=True) 
data.PTS.fillna(data.PTS.mean(),inplace=True) 
```

```python
data.isnull()
```

```python
#Tried to create a decision tree and replace null value with the predicting value, but encountered some errors
col=['G', 'GS', 'MP', 'PER']
notnull=data[pd.notnull(data.GS)]  
isnull=data[pd.isnull(data.GS)]  
from sklearn.ensemble import GradientBoostingRegressor  
G=GradientBoostingRegressor()  
G.fit(notnull[col].values,notnull.GS)  
isnull.GS=G.predict(isnull[col])  
data.GS[pd.isnull(data.GS)]=isnull.GS
```

```python
data.drop_duplicates()
```

```python
data.describe()
```

```python
#As for PER column, PER means the average points per game. So it cannot exceed 100 or be less than 0
data[data['PER']>100]
```

```python
data[data['PER']<0]
```

```python
#There is only one piece of data whose PER exceeds 100. So replace it with null.
data['PER']=data['PER'].replace([129.100000], np.nan)
```

```python
data.describe()
```

```python
data['PER']=data['PER'].replace([-90.600000], np.nan)
```

```python
#Replace all data which are less than 0 with null.
data['PER']=data['PER'].replace(data['PER'][data['PER']<0], np.nan)
```

```python
data.describe()
```

```python
#Replace new NA values with the average value of the column
data.PER.fillna(data.PER.mean(),inplace=True) 
```

```python
#Deal with TS data through the same method.
data['TS%']=data['TS%'].replace(data['TS%'][data['TS%']>1], data['TS%'][data['TS%']].mean())
```

```python
#Because of the dataset, the data all conforms to Normal Distribution
#Take some examples
#Age, the min Age is 18 and the max is 44
#Through the plot I could know that the main group of ages are between 25 and 30. With age decreasing on the left of 25, the possiblity decreses slower.
#On the other hand, with age increasing on the right of 30, the possiblity increses slower.
domain = np.arange(18, 44, 1)
values = stats.norm(data['Age'][data['Age']].mean(), 1).pdf(domain)
plt.plot(domain, values, color='r', linewidth=2)
plt.ylabel("Probability")
plt.title("Age Normal Distributions")
plt.show()
```

```python
#FT%, the min FT% is 0 and the max is 1
#Through the plot I could know that the main group of FT% are between 0.2 and 1. 
domain = np.arange(0, 1, 0.01)
values = stats.norm(data['FT%'][data['FT%']].mean(), 1).pdf(domain)
plt.plot(domain, values, color='r', linewidth=2)
plt.ylabel("Probability")
plt.title("FT% Normal Distributions")
plt.show()
```

```python
#The span of PTS data is very large
#From the formal statistics we know that the count of PTS is 18964
#Through the plot I could know that the main group of PTS are between 400 and 600. With PTS decreasing on the left of average point, the possiblity decreses slower.
#On the other hand, with PTS increasing on the right of average point, the possiblity increses slower. Because I think all my plots are about normal distribution and it's similar to normal distribution, and I want to change a plot here.
arrival_rate_1 = stats.poisson.rvs(size=18964,  
                                   mu=data['PTS'].mean() )       


# Plot histogram
pd.DataFrame(arrival_rate_1).hist(range=(-0.5,max(arrival_rate_1)+0.5)
                                    , bins=max(arrival_rate_1)+1)
```
